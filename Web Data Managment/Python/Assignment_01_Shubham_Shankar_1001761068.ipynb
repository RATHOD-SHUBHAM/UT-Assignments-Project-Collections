# Question 1:

   ### Logic: 
   - Program execution start from the main function. 
   - Once it the enter the main function. It request the user to enter three different input.

### Input and Output:

```python
def main():
    try:
        string_one = input("Enter string 1: ") # request user to enter the first string.
        string_two = input("Enter string 2: ") # request user to enter the second string.
        string_three = input("Enter string 3: ") # request user to enter the third string.
        
        str_concat = string_one + string_two    # concatenate string one and string two obtained from user
        
        if str_concat == string_three:      # if concatenated string ( string 1 + string 2) is equal to string 3 then  print.
            print("They are equal")
        else:
            reverse_str_concat = string_two + string_one    # if they are not qual then concatenate string(one and two) in reverse like:  string_two + string_one
            
            if reverse_str_concat == string_three:    #if concatenated string ( string 2 + string 1) is equal to string 3 then  print they are equal else not.
                print("They are also equal")
            else:
                print("They are not equal")
             
    except:
        print("Please enter a valid input !!")
```

```python
if __name__ == '__main__':
    main()    # call to the main function
```

```python
import math

def main():
    user_input = int(input("Enter set of 3 or 4 integers: "))  # request user to input the number of interger they want to perform operation upon.
        
    print("\n")
        
    try:
        
        if user_input == 3 or user_input == 4:  # user must enter only 3 or 4 integers
            
            # Block is entered only when user input is 3
            if user_input == 3:  # if user wants to enter 3 integer then do this 
                
                # block where user enters his 3 integer number
                first_number = int(input("Enter First Number: "))
                second_number = int(input("Enter Second Number: "))
                third_number = int(input("Enter Third Number: "))
                
                # block to calculate average number.
                average_number = (int(first_number + second_number + third_number) / 3) 
                
                if average_number % int(average_number) < 0.5:   # this will let us know if we have to round up or round down.
                    average_number = math.floor(average_number)   # if average is less than 0.5 then round down.
                else:
                    average_number = math.ceil(average_number)    # if average is greater than 0.5 then round up.
                    
                # block to calculate the median
                
                user_number = [first_number,second_number,third_number]  # converting the user input to a list of integer.
                user_number.sort()    # In order to calculate Median, Number should be in sorted order.
                
                
                median = int((len(user_number)-1)/2)
                
                
                print("\n")
                print("The average of the three number is : ",average_number)
                print("The maximum of the three number is: ",max(first_number,second_number,third_number))   # calculate the max of the three number.
                print("The minimum of the three number is: ",min(first_number,second_number,third_number))   # calculate the min of the three number.
                print("The median of the three number is : ",user_number[median])
                    
            # This block is entered when user input is 4
            else:     # if user wants to enter 4 integer then else part will be executed
                first_number = int(input("Enter First Number: "))
                second_number = int(input("Enter Second Number: "))
                third_number = int(input("Enter Third Number: "))
                fourth_number= int(input("Enter Fourth Number: "))
                
                # block to calculate average number.
                average_number = (int(first_number + second_number + third_number + fourth_number) / 4)
               
                if average_number % int(average_number) < 0.5:   # this will let us know if we have to round up or round down.
                    average_number = math.floor(average_number)   # if average is less than 0.5 then round down.
                else:
                    average_number = math.ceil(average_number)    # if average is greater than 0.5 then round up.
                    
                # block to calculate the median
                
                user_number = [first_number,second_number,third_number,fourth_number]  # converting the user input to a list of integer.
                user_number.sort()    # In order to calculate Median, Number should be in sorted order.
                
                mid_value_one = int(len(user_number)/2)           # finding the first mid point
                mid_value_two = int(len(user_number)/2) - 1      # finding the second mid  point
                median = ((user_number[mid_value_one]) + (user_number[mid_value_two]))/2
                
                
                
                if median % int(median) < 0.5:   # this will let us know if we have to round up or round down.
                    median = math.floor(median)   # if median is less than 0.5 then round down.
                else:
                    median = math.ceil(median)   # if median is greater than  or equal 0.5 then round up.
                
                
                
                print("\n")
                print("The average of the three number is : ",average_number)
                print("The maximum of the three number is: ",max(first_number,second_number,third_number,fourth_number))    # calculate the max of the three number.
                print("The minimum of the three number is: ",min(first_number,second_number,third_number,fourth_number))    # calculate the min of the three number.
                print("The median of the three number is : ",median)
                    
        else:
            print("Invalid Input")
            print("Please enter either 3 or 4 only !! ")
        
        
    except:
        print("Enter a valid input!! Make sure it is a Valid Interger Number")

```

```python
if __name__ == '__main__':
    main()
```

# Question 3:

   ### Logic: 
   - Program execution start from the main function. 
   - Once it the enter the main function. It request user to enter a positive integer.
   - if the user enters 0 or 1 then: 
       - program identifies it as not prime or composite.
   - if the user enters number greater than 1: 
       - Then the program check if the number is divisible bby itself or no. 
       - If it is divisible by one or itself it returns as prime. Else it returns as composite number.
       
   - Note: Since the program takes user input continueously. The user needs to stop the program execution explicitely.
   
   ### Input And Output :
   
        Enter positive integer: 5

        It is a prime

        Enter positive integer: 6

        It is a composite

        Enter positive integer: 0

        It is neither prime nor composite

        Enter positive integer: -2

        Error: You did not enter a positive integer

        Enter positive integer: 2.5

        Error: You did not enter a positive integer

        Enter positive integer: what is integer?

        Error: You did not enter a positive integer

 

```python
def main():
    while(True):
        user_input = int(input("Enter a positive number: "))   # request user to enter a positive number
        try:
            if user_input == 0 or user_input == 1:
                print("It is neither prime nor composite")   # if the user enters either 0 or 1.
                print("\n")
            elif user_input > 1:                             # check if the number is greater than 1.
                for i in range(2,user_input):       # check if the number is not divisible by any number
                    if (user_input % i) == 0:
                        print("It is a composite Number")
                        print("\n")
                        break
                else:
                    print("It is a prime")
                    print("\n")
            else:
                print("You did not enter a positive integer")
                print("\n")

        except:
            print("You did not enter a positive integer")     # if user input is not positve then this error will be shown
            print("\n")

```

```python
if __name__ =='__main__':
    main()
```

# Question 4:

   ### Logic: 
   - Program execution start from the main function. 
   - Once it the enter the main function. It request the user to enter two integer input. 
   - Once the program has all the three input. It check if the second number is zero.
       - If Zero, Program says division is not possible
       - Else, It calls the SmartDivision function
       
   - The SmartDivision function takes in two input that user gave. 
       - It then checks if on dividing input one with input two provides a remainder or not.
       - If there is no remainder, then Input one is greater than input two. So on division it return a integer value.
       - If it provides a remainder than, Input one is smaller than input two. So on division it return a floating value.

```python
def SmartDivision(user_input_one,user_input_two):
    try:
        if user_input_one % user_input_two == 0:               # if the remainder is zero then first number is divisible by second number
            result = int(user_input_one / user_input_two)
            result_type = type(result)
            print ("The result is: ",result)
            print("The result type is: ",result_type)
        else:
            result = user_input_one / user_input_two         # if remainder is not zero then there is some float value present
            result_type = type(result)
            print ("The result is: ",result)
            print("The result type is: ",result_type)
    except:
        print("Please provide a Valid Input")
```

```python
def main():
    try:
        user_input_one = int(input("Enter the first number: "))       # takes integer from user
        user_input_two = int(input("Enter the second number: "))
        print("\n")
        
        if user_input_two == 0:                                  # if the second integer is zero then division is not possible.
            print("You cannot divide a number by Zero")
        else:
            SmartDivision(user_input_one,user_input_two)       # if both the number are valid then make a function call
    except:
        print("Please provide a Valid Input")
            
```

```python
if __name__ == '__main__':
    main()
```
